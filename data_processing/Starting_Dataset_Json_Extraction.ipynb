### III. Extract eye gazing features from json files

```python
all_confidences = []
all_eye_directions = []
df['total_confidence'] = 0
df['number_of_frames_with_face'] = 0

print(df['number_of_frames_with_face'])
counter = 0
for i in range(len(df)):
    print(counter)
    number_of_frames_with_face = 0
    number_of_frames = 0
    total_confidence = 0
    confidences = []
    eye_directions = []

    video_key = df.iloc[i]['video_key']
    print(f'Processing video : {video_key}')
    
    folder_path = 'Downsampling/Rekognition_Downsampling/' + video_key + '/'
    # List all JSON files in the specified path in the S3 bucket
    json_files = client.list_objects(Bucket=bucket_name, Prefix=folder_path)['Contents']
    df.at[i, 'number_of_frames'] = len(json_files)
    # Process each frame
    for json_file in json_files:
        # Read the JSON data from S3
        json_data = client.get_object(Bucket=bucket_name, Key=json_file['Key'])
        json_content = json_data['Body'].read().decode('utf-8')
        data = json.loads(json_content)
        if 'FaceDetails' in data and len(data['FaceDetails']) > 0:
            face_details = data['FaceDetails'][0]
            yaw = face_details['EyeDirection']['Yaw']
            pitch = face_details['EyeDirection']['Pitch']
            confidence = face_details['EyeDirection']['Confidence']
            number_of_frames_with_face += 1
            total_confidence += confidence
            confidences.append(confidence)
            eye_directions.append([yaw, pitch])
        else:
            #print(f'No face detected in frame : {json_file["Key"]}')
            eye_directions.append([None, None])
            confidences.append(None)
        
    #We keep track of the number of frames extracted and the total confidence
    df.loc[i, 'number_of_frames_with_face'] = number_of_frames_with_face
    df.loc[i, 'total_confidence'] = total_confidence/number_of_frames_with_face
    all_confidences.append(confidences)
    all_eye_directions.append(eye_directions)
    counter += 1    
```

```python
df['confidences'] = all_confidences
df['eye_directions'] = all_eye_directions
```

IV. We run some sanity checks.

```python
#We do some sanity checks

#We check that the number of frames is the length of the list of confidences
for i in range(len(df)):
    assert len(df.iloc[i]['confidences']) == df.iloc[i]['number_of_frames']

#We check that the length of the list of eye_directions is the same as the length of the list of confidences
for i in range(len(df)):
    assert len(df.iloc[i]['eye_directions']) == len(df.iloc[i]['confidences'])

#We check that the number of frames with face is the length of the list of confidences without None values
for i in range(len(df)):
    assert df.iloc[i]['number_of_frames_with_face'] == len([x for x in df.iloc[i]['confidences'] if x is not None])

#We check that the total confidence is the sum of the confidences divided by the number of frames with face
for i in range(len(df)):
    assert df.iloc[i]['total_confidence'] == sum([x for x in df.iloc[i]['confidences'] if x is not None])/df.iloc[i]['number_of_frames_with_face']
```

## IV. Extract Blinking information from Json Files

```python
df = pd.read_csv('starting_dataset_rekognition_extracted_tested.csv')
df
```

```python
df
```

```python
counter = 0

for i in range(238, len(df)):
    print(counter)
    blinks = 0
    previous_eye = True
    frames_with_face = 0
    video_key = df.iloc[i]['video_key']
    print(f'Processing video : {video_key}')
    
    folder_path = 'Downsampling/Rekognition_Downsampling/' + video_key + '/'
    # List all JSON files in the specified path in the S3 bucket
    json_files = client.list_objects(Bucket=bucket_name, Prefix=folder_path)['Contents']
    df.at[i, 'number_of_frames'] = len(json_files)
    # Process each frame
    for json_file in json_files:
        # Read the JSON data from S3
        json_data = client.get_object(Bucket=bucket_name, Key=json_file['Key'])
        json_content = json_data['Body'].read().decode('utf-8')
        data = json.loads(json_content)
        if 'FaceDetails' in data and len(data['FaceDetails']) > 0:
            face_details = data['FaceDetails'][0]
            open_eye = face_details['EyesOpen']['Value']
            confidence = face_details['EyesOpen']['Confidence']
            frames_with_face += 1
            if confidence > 80:
                if open_eye:
                    previous_eye = True
                else:
                    if previous_eye == True:
                        blinks += 1
                    previous_eye = False        
        
    fps = df.iloc[i]['fps']
    face_duration = frames_with_face/fps
    blinkspermin = blinks/(face_duration/60)
    df.loc[i, 'blinking_rate'] = blinkspermin
    print(blinkspermin)
    counter += 1    
```
