```python
df_nt = df[df['ASD'] == 0]
df_nt
```

```python
df_nt['video_key']
```

```python
#We create a list of all the videos
video_list = df_nt['video_key'].tolist()
video_list
```

```python
len(video_list)
```

### We download the videos.

```python
#We want to play all the videos in the dataset to see if we keep them or not

download = False

if download : 
    #We connect to the bucket
    bucket = resource.Bucket(bucket_name)

    # We create a folder to store the videos
    if not os.path.exists('videos'):
        os.makedirs('videos')

    # We download all the videos in the folder
    counter = 0
    for video in video_list:
        try:
            bucket.download_file(video, "videos/{}.mp4".format(counter))
            counter += 1
        except Exception as e:
            print(f"Error downloading {video}: {e}")
```

### We look at the videos and exclude the videos deleted. 

```python
#We look at folder videos
videos_kept = []

for filename in os.listdir('videos'):
    if filename.endswith(".mp4"):
        index = int(filename.split('.')[0])
        videos_kept.append(index)
    else:
        continue

len(videos_kept)
```

```python
len(video_list)
```

```python
#We only keep the indexes in video_list
video_list_kept = [video_list[i] for i in videos_kept]
#We look the videos we didnt keep (ie. the videos in video_list that are not in video_list_kept)
video_list_not_kept = [video for video in video_list if video not in video_list_kept]
video_list_not_kept
```

```python
#We add a column flagging the videos we keep (0 if we don't keep it, 1 if we keep it) in 

df['flag'] = 1

#We set the flag to 0 for videos in video_list_not_kept
for video in video_list_not_kept:
    df.loc[df['video_key'] == video, 'flag'] = 0

df['flag'].value_counts()
```

```python
df.to_csv('starting_dataset_rekognition_extracted_tested_with_flag.csv', index=False)
```
